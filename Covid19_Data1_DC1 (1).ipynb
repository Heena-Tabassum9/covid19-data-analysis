```python
import pandas as pd
import numpy as np
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load and store dataframe in variable

covid19_df = pd.read_csv('COVID-19_Dataset1.csv')
```

```python
# Printing Dataframe or shows CSV file

covid19_df
```

```python
# .info() function will give the basic info about the dataset

covid19_df.info()
```

```python
# .describe() shows the statistical description of dataset

covid19_df.describe()
```

```python
# Printing Dataframe or shows CSV file
covid19_df.head()
```

```python
# Creating new Required columns in dataframe by using .apply()

covid19_df['TotalConfirmed'] = covid19_df.apply(lambda row: row['Confirmed'] + row['newConfirmed'], axis=1)
covid19_df['TotalDeath'] = covid19_df.apply(lambda row: row['Death'] + row['newDeath'], axis=1)
covid19_df['Recovered'] = covid19_df.apply(lambda row: row['TotalConfirmed'] - row['TotalDeath'], axis=1)
```

```python
covid19_df
```

```python
# .drop() unnamed Column dont need anymore so drop it



columnsToRemove = ['Unnamed: 0', 'Confirmed', 'newConfirmed', 'Death', 'newDeath']

covid19_df = covid19_df.drop(columns=columnsToRemove, axis=1)
```

```python
# printing for confirming columns are deleted

covid19_df
```

```python
# Checking Dtype of Column Country 

covid19_df['Country']
```

```python
# Chenging Dtype of Column Country from obvject to String

covid19_df['Country'] = covid19_df['Country'].convert_dtypes()
```

```python
# Printing dataframe for confirming Dtype changed

covid19_df['Country']
```

```python
# Checking Dtype of Column Date 

covid19_df['Date']
```

```python
# now first we need to convert dtype of country object to string.


covid19_df['Date'] = pd.to_datetime(covid19_df['Date'])
```

```python
# Printing dataframe for confirming Dtype changed

covid19_df['Date']
```

```python
# Dates having 0 values to remove

dates_to_remove = pd.to_datetime(['12/31/2019'])


# Removing rows with specified dates

covid19_df = covid19_df[~covid19_df['Date'].isin(dates_to_remove)]
print(covid19_df)
```

```python
# Cheqking whether the date of 31/12/2019 removed or not?

covid19_df.head(15)
```

```python
# Given dataset having daily reports. need to create & convert Date(DD-MM-YYYY) column in Month(YYYY-MM)


covid19_df['Month'] = covid19_df['Date'].dt.to_period('M')
```

```python
# printing for com=nfirming change in dataset

covid19_df.head(15)
```

```python
# Now we need to by country and Date


covid19_df = covid19_df.groupby(['Country', "Month"]).agg(
    days_inMonth=('Month', 'count'),
    total_confirmed=('TotalConfirmed', 'sum'),
    total_deaths=('TotalDeath', 'sum'),
    total_recoverd=('Recovered', 'sum'),
    
).reset_index()



covid19_df
```

```python
#Saving cleaned dataframe in csv format

covid19_df.to_csv('COVID19_CleanDataset1.csv', index=False)
```

```python
'''
Z-score Method:
This method identifies outliers based on how many standard deviations away from the mean a data point lies.
'''
# Z-score Method for 'total_confirmed' Column

from scipy.stats import zscore

covid19_df2['z_score'] = zscore(covid19_df2['total_confirmed'])
threshold = 2  # Common threshold for Z-score

df_cleaned_zscore = covid19_df2[abs(covid19_df2['z_score']) < threshold].drop(columns=['z_score'])
```

```python
# Z-score Method for 'total_deaths' Column


covid19_df2['z_score'] = zscore(covid19_df2['total_deaths'])
threshold = 2  # Common threshold for Z-score

df_cleaned_zscore = covid19_df2[abs(covid19_df2['z_score']) < threshold].drop(columns=['z_score'])
```
